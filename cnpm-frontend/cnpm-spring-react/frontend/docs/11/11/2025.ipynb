## Search Implementation Details

### How It Works:

1. **User types** in search box
2. **State updates** (`setSearchTerm`)
3. **Filter runs** on every keystroke
4. **Table updates** with filtered data
5. **Count displays** number of results

### Search Algorithm:

```javascript
const search = searchTerm.toLowerCase();
return (
  household.soHoKhau?.toLowerCase().includes(search) ||
  household.tenChuHo?.toLowerCase().includes(search) ||
  household.diaChi?.toLowerCase().includes(search)
);
```

**Features:**
- **Case-insensitive**: `toLowerCase()`
- **Partial match**: `includes()`
- **Safe access**: `?.` optional chaining
